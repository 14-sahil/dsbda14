```python
import requests
import bs4
```

```python
request1 = requests.get("https://www.etsy.com/ca/listing/1375008231/aluminum-composite-panel-dibond-acp-3mm?ga_order=most_relevant&ga_search_type=all&ga_view_type=gallery&ga_search_query=aluminum&ref=sc_gallery-1-1&sts=1&plkey=f196afa4b937468dcc195bf9bb59b383262c3236%3A1375008231")
request1

```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
# Fetching Reviews/Comments
reviews = soup.findAll('p', {'class': 'wt-text-truncate--multi-line wt-break-word wt-text-body'})
for review in reviews:
    print(review.get_text() + '\n\n')
```

```python
# Average overall rating
ratings = soup.find('span', {'class': 'wt-display-inline-block wt-mr-xs-1'})
ratings('input')[1]['value']
```

```python
# Individual ratings
individual_ratings = soup.findAll('span', {'class': 'wt-display-inline-block wt-mr-xs-1'})

for i in individual_ratings:
    print(i.get_text() + '\n')
```

```python
customer = soup.findAll('a', {'class': 'wt-text-link wt-mr-xs-1'})
for cust in customer:
    print(cust.get_text() + '\n')
```
