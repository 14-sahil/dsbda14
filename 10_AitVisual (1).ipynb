```python
# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Year', y='NH3', data=df)
plt.title('NH3 Levels by Year')
plt.xlabel('Year')
plt.ylabel('NH3')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['O3'], bins=20, color='skyblue', edgecolor='black')
plt.title('O3 Distribution')
plt.xlabel('O3')
plt.ylabel('Frequency')
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Year', y='CO', data=df)
plt.title('CO Levels by Year')
plt.xlabel('Year')
plt.ylabel('CO')
plt.xticks(rotation=45)
plt.show()
```

```python
# Timeseries chart
df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='NH3', data=df)
plt.title('NH3 Over MONTH')
plt.xlabel('MONTH')
plt.ylabel('NH3')
plt.xticks(rotation=45)
plt.show()
```
