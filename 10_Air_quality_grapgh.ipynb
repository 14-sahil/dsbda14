```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset
df = pd.read_csv(r'C:\Users\Sahil Karamkar\Downloads\DSBDA CHITS WISE CODES\Visualization.csv', delimiter=',')
df
```

```python
# Line chart with different x and y variables
plt.figure(figsize=(10, 6))
plt.plot(df['City'], df['PM10'])
plt.title('PM10 Levels Across Cities')
plt.xlabel('City')
plt.ylabel('PM10')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Fill missing values with zeros
df.fillna(0, inplace=True)

# Barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='PM10', data=df)
plt.title('PM10 Levels by City')
plt.xlabel('City')
plt.ylabel('PM10')
plt.xticks(rotation=45)
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='NO2', y='SO2', data=df)
plt.title('NO2 vs SO2')
plt.xlabel('NO2')
plt.ylabel('SO2')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['O3'], bins=20, color='skyblue', edgecolor='black')
plt.title('O3 Distribution')
plt.xlabel('O3')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='City', y='NH3', data=df)
plt.title('NH3 Levels by City')
plt.xlabel('City')
plt.ylabel('NH3')
plt.xticks(rotation=45)
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='City', y='CO', data=df)
plt.title('CO Levels by City')
plt.xlabel('City')
plt.ylabel('CO')
plt.xticks(rotation=45)
plt.show()
```

```python
# Violin plot 
plt.figure(figsize=(10, 6))
sns.violinplot(x='AQI_Bucket', y='PM10', data=df)
plt.title('PM10 Levels by AQI Bucket')
plt.xlabel('AQI Bucket')
plt.ylabel('PM10')
plt.xticks(rotation=45)
plt.show()
```

```python
# Timeseries chart
df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='AQI', data=df)
plt.title('AQI Over Time')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.xticks(rotation=45)
plt.show()
```
