```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
```

```python
df
```

# line chart

```python
df.shape
```

```python
plt.figure(figsize=(10, 6))
plt.title('Line Chart')
sns.lineplot(x=df['age'], y=df['chol'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Bar Chart')
plt.xlabel('Age')
plt.ylabel('Chol')
plt.bar(df['age'], df['chol'], color='red')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Bar Chart')
sns.barplot(x=df['output'], y=df['age'])
plt.show()
```

# heatmap

```python
corr = df.corr()
corr
```

```python
plt.figure(figsize=(10, 6))
plt.title('Heatmap')
sns.heatmap(corr, annot=True)
plt.show()
```

# scatter plot

```python
plt.figure(figsize=(10, 6))
plt.title('Scatter Plot')
sns.scatterplot(x=df['age'], y=df['chol'])
plt.show()
```

# histogram

```python
plt.title('Histogram')
sns.histplot(x=df['age'])
plt.show()
```

# boxplot

```python
plt.figure(figsize=(10, 6))
plt.title('Boxplot')
sns.boxplot(y=df['age'])
plt.show()
```

# violin plot

```python
plt.figure(figsize=(10, 6))
plt.title('Violin Plot')
sns.violinplot(y=df['age'])
plt.show()
```
