```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tips_DV.csv')
```

```python
df
```

```python
df.rename(columns = {'size':'peoples'}, inplace = True)
```

```python
df.isna().sum()
```

```python
col_var = df[['peoples','sex', 'smoker', 'day', 'time']]

for i in col_var:
    print(i, df[i].unique())
```

```python
df.columns
```

## Plots

```python
# Select numeric columns only for correlation calculation
numeric_cols = df.select_dtypes(include=['number'])


# Create heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.show()
```

## Line Plots

```python
plt.figure(figsize=(10, 6))
plt.title('Line Plot: Tip vs Total Bill')
sns.lineplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Line Plot: Tip vs Number of People')
sns.lineplot(x='peoples', y='tip', data=df)
plt.xlabel('Number of People')
plt.ylabel('Tip')
plt.show()
```

## Box Plots

```python
plt.figure(figsize=(10, 6))
plt.title('Box Plot: Total Bill by Day')
sns.boxplot(x='day', y='total_bill', data=df)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Box Plot: Total Bill by Day and Sex')
sns.boxplot(x='day', y='total_bill', hue='sex', data=df)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

## Histogram

```python
plt.figure(figsize=(10, 6))
plt.title('Histogram: Total Bill by Time')
sns.histplot(x='total_bill', hue='time', data=df)
plt.xlabel('Total Bill')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Histogram: Tip by Sex')
sns.histplot(x='tip', hue='sex', data=df)
plt.xlabel('Tip')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Histogram: Tip by Smoker')
sns.histplot(x='tip', hue='smoker', data=df)
plt.xlabel('Tip')
plt.show()
```

## Scatter Plots

```python
plt.figure(figsize=(10, 6))
plt.title('Scatter Plot: Total Bill vs Tip (Time-wise)')
sns.scatterplot(x='total_bill', y='tip', hue='time', data=df)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

## Violen

```python
plt.figure(figsize=(10, 6))
plt.title('Violin Plot: Tip by Day')
sns.violinplot(x='day', y='tip', data=df)
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Violin Plot: Total Bill by Number of People')
sns.violinplot(x='peoples', y='total_bill', data=df)
plt.xlabel('Number of People')
plt.ylabel('Total Bill')
plt.show()
```

# Bar plot

```python

plt.figure(figsize=(10, 6))
plt.title('Bar Plot')
sns.barplot(x='day', y='total_bill', data=df)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()

```

# Time series chart

```python

plt.figure(figsize=(10, 6))
plt.title('Time Series Chart')
sns.lineplot(x='peoples', y='tip', data=df)
plt.xlabel('Number of People')
plt.ylabel('Tip')
plt.show()
```
